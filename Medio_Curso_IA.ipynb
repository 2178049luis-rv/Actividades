Parte 1: Carga y Limpieza (Pandas y Numpy)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv"
df = pd.read_csv(url)
df.head()
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df['horsepower'] = df['horsepower'].fillna(df['horsepower'].median())
df.isnull().sum()
```

```python
df['relacion_peso_potencia'] = df['weight'] / df['horsepower']
df.head()
```

Parte 2: Visualización de Datos (Matplotlib)

```python
plt.figure()
plt.hist(df['acceleration'], bins=15)
plt.title('Distribución de la Aceleración')
plt.xlabel('Aceleración')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.figure()
plt.scatter(df['displacement'], df['horsepower'])
plt.title('Desplazamiento vs Caballos de Fuerza')
plt.xlabel('Desplazamiento (displacement)')
plt.ylabel('Caballos de Fuerza (horsepower)')
plt.show()
```

```python
promedio_mpg = df.groupby('origin')['mpg'].mean()
plt.figure()
promedio_mpg.plot(kind='bar')
plt.title('Promedio de MPG por Origen')
plt.xlabel('Origen')
plt.ylabel('MPG Promedio')
plt.show()
```
